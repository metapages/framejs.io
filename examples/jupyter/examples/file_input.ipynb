# Load a File and Pass to Metaframe Widget

Read a local file in Python and push its content as inputs to a metaframe widget.

```python
from metaframe_widget import MetaframeWidget
```

## 1. Load the file

```python
import csv

with open("data.csv") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

print(f"Loaded {len(rows)} rows")
rows[:3]
```

## 2. Create a widget that renders the data as a table

```python
# A table-viewer metaframe (built in the framejs.io editor, saved to a short URL):
# it renders inputs.rows as an HTML table and echoes the row count.
viewer = MetaframeWidget(
    url="https://framejs.io/j/152de7cc6e1d16cfb957dafec4422ce845cbcb42ae9f1fdf956ffc7b2496fbe7",
    height="300px",
)

viewer
```

## 3. Pass the file content to the widget

```python
viewer.set_inputs({"rows": rows})
```

```python
# Outputs arrive asynchronously — re-run this cell after a moment if it shows {}
viewer.outputs
```
